```python
#List top 5 milk producing states in each year from 2010 to 2015
df2["2011"] =(df2["Cow Milk-2010-11"] + df2["Boffalo Milk-2010-11"] + df2["Goat Milk-2010-11"]).sort_values(ascending=False)
df2["2012"]= df2["Cow Milk-2011-12"]  + df2["Goat Milk-2011-12"]
df2["2013"]= df2["Cow Milk-2013-14"] + df2["Boffalo Milk-2013-14"] + df2["Goat Milk-2013-14"]
df2["2014"]= df2["Cow Milk-2014-15"] + df2["Boffalo Milk-2014-15"] + df2["Goat Milk-2014-15"]
df2["2015"]= df2["Cow Milk-2015-16"] + df2["Boffalo Milk-2015-16"] + df2["Goat Milk-2015-16"]
df2["all"]=df2["2011"]+df2["2012"]+df2["2013"]+df2["2014"]+df2["2015"]
print(df2.sort_values('all',ascending=False)['State/ UT Name'].head())
```

```python
#3.Calculate average milk production by cow, buffalo and goat in each year for selected state.
df2["cow_allyear"]=df2["Cow Milk-2010-11"]+df2["Cow Milk-2011-12"]+df2["Cow Milk-2013-14"]+df2["Cow Milk-2014-15"]+df2["Cow Milk-2015-16"]
df2["boffalo_allyear"]=df2["Boffalo Milk-2010-11"]+df2["Boffalo Milk-2011-12"]+df2["Boffalo Milk-2013-14"]+df2["Boffalo Milk-2014-15"]+df2["Boffalo Milk-2015-16"]
df2["goat_allyear"]=df2["Goat Milk-2010-11"]+df2["Goat Milk-2011-12"]+df2["Goat Milk-2013-14"]+df2["Goat Milk-2014-15"]+df2["Goat Milk-2015-16"]
df2["total"]=df2["cow_allyear"]+df2["boffalo_allyear"]+df2["goat_allyear"]
df2[df2["total"].mean() and df2["State/ UT Name"]=="Uttar Pradesh"]
```
